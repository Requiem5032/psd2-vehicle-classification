```python
import os
from os.path import dirname

root_dir = dirname(os.getcwd())
os.chdir(root_dir)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bike_df = pd.read_csv('data/bike.csv')
bike_ax = bike_df['ax'].to_numpy()
bike_ay = bike_df['ay'].to_numpy()
bike_az = bike_df['az'].to_numpy()
```

```python
bus_df = pd.read_csv('data/bus.csv')
bus_ax = bus_df['ax'].to_numpy()
bus_ay = bus_df['ay'].to_numpy()
bus_az = bus_df['az'].to_numpy()
```

```python
rail_df = pd.read_csv('data/rail.csv')
rail_ax = rail_df['ax'].to_numpy()
rail_ay = rail_df['ay'].to_numpy()
rail_az = rail_df['az'].to_numpy()
```

```python
car_df = pd.read_csv('data/car-clear.csv')
car_ax = car_df['ax'].to_numpy()
car_ay = car_df['ay'].to_numpy()
car_az = car_df['az'].to_numpy()
```

```python
fig1, (ax1_fig1, ax2_fig1, ax3_fig1) = plt.subplots(3, 1, figsize=(10, 6))
ax1_fig1.plot(bike_ax)
ax2_fig1.plot(bike_ay)
ax3_fig1.plot(bike_az)
ax1_fig1.set_xlabel('Data points')
ax2_fig1.set_xlabel('Data points')
ax3_fig1.set_xlabel('Data points')
ax1_fig1.set_title('X-axis')
ax2_fig1.set_title('Y-axis')
ax3_fig1.set_title('Z-axis')
fig1.suptitle('Bike Accelerometer Data')
fig1.tight_layout()
plt.show()
```

```python
fig2, (ax1_fig2, ax2_fig2, ax3_fig2) = plt.subplots(3, 1, figsize=(10, 6))
ax1_fig2.plot(bus_ax)
ax2_fig2.plot(bus_ay)
ax3_fig2.plot(bus_az)
ax1_fig2.set_xlabel('Data points')
ax2_fig2.set_xlabel('Data points')
ax3_fig2.set_xlabel('Data points')
ax1_fig2.set_title('X-axis')
ax2_fig2.set_title('Y-axis')
ax3_fig2.set_title('Z-axis')
fig2.suptitle('Bus Accelerometer Data')
fig2.tight_layout()
plt.show()
```

```python
fig3, (ax1_fig3, ax2_fig3, ax3_fig3) = plt.subplots(3, 1, figsize=(10, 6))
ax1_fig3.plot(rail_ax)
ax2_fig3.plot(rail_ay)
ax3_fig3.plot(rail_az)
ax1_fig3.set_xlabel('Data points')
ax2_fig3.set_xlabel('Data points')
ax3_fig3.set_xlabel('Data points')
ax1_fig3.set_title('X-axis')
ax2_fig3.set_title('Y-axis')
ax3_fig3.set_title('Z-axis')
fig3.suptitle('Rail Accelerometer Data')
fig3.tight_layout()
plt.show()
```

```python
fig4, (ax1_fig4, ax2_fig4, ax3_fig4) = plt.subplots(3, 1, figsize=(10, 6))
ax1_fig4.plot(car_ax)
ax2_fig4.plot(car_ay)
ax3_fig4.plot(car_az)
ax1_fig4.set_xlabel('Data points')
ax2_fig4.set_xlabel('Data points')
ax3_fig4.set_xlabel('Data points')
ax1_fig4.set_title('X-axis')
ax2_fig4.set_title('Y-axis')
ax3_fig4.set_title('Z-axis')
fig4.suptitle('Car Accelerometer Data')
fig4.tight_layout()
plt.show()
```

```python
from scipy import stats
```

```python
bike_ax_kde = stats.gaussian_kde(bike_ax)
bike_ay_kde = stats.gaussian_kde(bike_ay)
bike_az_kde = stats.gaussian_kde(bike_az)
bus_ax_kde = stats.gaussian_kde(bus_ax)
bus_ay_kde = stats.gaussian_kde(bus_ay)
bus_az_kde = stats.gaussian_kde(bus_az)
rail_ax_kde = stats.gaussian_kde(rail_ax)
rail_ay_kde = stats.gaussian_kde(rail_ay)
rail_az_kde = stats.gaussian_kde(rail_az)
car_ax_kde = stats.gaussian_kde(car_ax)
car_ay_kde = stats.gaussian_kde(car_ay)
car_az_kde = stats.gaussian_kde(car_az)
```

```python
x_eval = np.linspace(-10, 10, 200)
```

```python
fig5, (ax1_fig5, ax2_fig5, ax3_fig5) = plt.subplots(3, 1, figsize=(10, 6))
ax1_fig5.plot(x_eval, bike_ax_kde(x_eval))
ax2_fig5.plot(x_eval, bike_ay_kde(x_eval))
ax3_fig5.plot(x_eval, bike_az_kde(x_eval))
ax1_fig5.plot(bike_ax, np.zeros_like(bike_ax), 'r+', ms=20)
ax2_fig5.plot(bike_ay, np.zeros_like(bike_ay), 'r+', ms=20)
ax3_fig5.plot(bike_az, np.zeros_like(bike_az), 'r+', ms=20)
ax1_fig5.set_title('X-axis')
ax2_fig5.set_title('Y-axis')
ax3_fig5.set_title('Z-axis')
fig5.suptitle('Bike Accelerometer Kernel Density Estimation (KDE)')
fig5.tight_layout()
plt.show()
```

```python
fig6, (ax1_fig6, ax2_fig6, ax3_fig6) = plt.subplots(3, 1, figsize=(10, 6))
ax1_fig6.plot(x_eval, bus_ax_kde(x_eval))
ax2_fig6.plot(x_eval, bus_ay_kde(x_eval))
ax3_fig6.plot(x_eval, bus_az_kde(x_eval))
ax1_fig6.plot(bus_ax, np.zeros_like(bus_ax), 'r+', ms=20)
ax2_fig6.plot(bus_ay, np.zeros_like(bus_ay), 'r+', ms=20)
ax3_fig6.plot(bus_az, np.zeros_like(bus_az), 'r+', ms=20)
ax1_fig6.set_title('X-axis')
ax2_fig6.set_title('Y-axis')
ax3_fig6.set_title('Z-axis')
fig6.suptitle('Bus Accelerometer Kernel Density Estimation (KDE)')
fig6.tight_layout()
plt.show()
```

```python
fig7, (ax1_fig7, ax2_fig7, ax3_fig7) = plt.subplots(3, 1, figsize=(10, 6))
ax1_fig7.plot(x_eval, rail_ax_kde(x_eval))
ax2_fig7.plot(x_eval, rail_ay_kde(x_eval))
ax3_fig7.plot(x_eval, rail_az_kde(x_eval))
ax1_fig7.plot(rail_ax, np.zeros_like(rail_ax), 'r+', ms=20)
ax2_fig7.plot(rail_ay, np.zeros_like(rail_ay), 'r+', ms=20)
ax3_fig7.plot(rail_az, np.zeros_like(rail_az), 'r+', ms=20)
ax1_fig7.set_title('X-axis')
ax2_fig7.set_title('Y-axis')
ax3_fig7.set_title('Z-axis')
fig7.suptitle('Rail Accelerometer Kernel Density Estimation (KDE)')
fig7.tight_layout()
plt.show()
```

```python
fig8, (ax1_fig8, ax2_fig8, ax3_fig8) = plt.subplots(3, 1, figsize=(10, 6))
ax1_fig8.plot(x_eval, car_ax_kde(x_eval))
ax2_fig8.plot(x_eval, car_ay_kde(x_eval))
ax3_fig8.plot(x_eval, car_az_kde(x_eval))
ax1_fig8.plot(car_ax, np.zeros_like(car_ax), 'r+', ms=20)
ax2_fig8.plot(car_ay, np.zeros_like(car_ay), 'r+', ms=20)
ax3_fig8.plot(car_az, np.zeros_like(car_az), 'r+', ms=20)
ax1_fig8.set_title('X-axis')
ax2_fig8.set_title('Y-axis')
ax3_fig8.set_title('Z-axis')
fig8.suptitle('Car Accelerometer Kernel Density Estimation (KDE)')
fig8.tight_layout()
plt.show()
```
